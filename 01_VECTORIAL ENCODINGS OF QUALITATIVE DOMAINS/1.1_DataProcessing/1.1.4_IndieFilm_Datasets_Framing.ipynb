```python
import os
import cv2
import random

video_folder = r'D:\FinalAssignment\Movies'
keyframes_folder = r'D:\FinalAssignment\MoviesKeyframes'

if not os.path.exists(keyframes_folder):
    os.makedirs(keyframes_folder)

for root, dirs, files in os.walk(video_folder):
    for file in files:
        video_path = os.path.join(root, file)
        video_name = os.path.splitext(file)[0]

        cap = cv2.VideoCapture(video_path)
        frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

        if frame_count < 20:
            print(f"The frame amount of the video {video_name} is not enough, thus cannot be framed")
            continue

        selected_frames = random.sample(range(frame_count), 20)

        for frame_number in selected_frames:
            cap.set(cv2.CAP_PROP_POS_FRAMES, frame_number)
            ret, frame = cap.read()
            if ret:
                frame_save_path = os.path.join(keyframes_folder, f"{video_name}_frame_{frame_number}.jpg")
                cv2.imwrite(frame_save_path, frame)

        print(f"Video {video_name} framing done")

        cap.release()
```
